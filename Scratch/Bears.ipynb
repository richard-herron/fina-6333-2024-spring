```python
import pandas as pd
import yfinance as yf
```

```python
raw = yf.download('^GSPC')
```

```python
df = (
    raw
    .iloc[:-1]
    .assign(
        top=lambda x: x['Close'].cummax(),
        dd=lambda x: x['Close'] / x['top'] - 1,
        top_n=lambda x: (x['Close']==x['top']).cumsum(),
        dd_max=lambda x: x.groupby('top_n')['dd'].transform('min'),
        dd_max_date=lambda x: x.groupby('top_n')['dd'].transform('idxmin')
    )
    .sort_values('dd_max')
)

df
```
